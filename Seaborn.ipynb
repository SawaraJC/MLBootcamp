Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
#heat map (data must be in matrix form i.e. columns and rows should be relevant)
tc = tips.corr()
```

```python
sns.heatmap(tc, annot= True)   #how two values are correlated to each other
```

```python
tc
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='magma')
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
sns.clustermap(fp, standard_scale=1)
```

Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(data=iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time',row='smoker')
g.map(sns.distplot, 'total_bill')
```
