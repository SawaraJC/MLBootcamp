```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#allows us to see all the visualisation inside the notebook
```

```python
%matplotlib inline  
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional method
```

```python
plt.plot(x, y, 'g-')  #third argument for color of graph

#adding x and y label 
plt.xlabel('X-array')
plt.ylabel('Y-array')

#add title to the graph
plt.title('Practice graph')

plt.show()
```

```python
#multiple plots on same canvas
plt.subplot(1,2,1)    #arguments: rows, columns, plot no.
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#object oriented method
fig = plt.figure()   #creating object of figure class

axes = fig.add_axes([0.1,0.1,0.8,0.8])   #four arguments: left, bottom, height, width
axes.plot(x,y, 'b')

axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#subplots using object oriented method
fig, axes = plt.subplots(nrows=1, ncols=2)

#taking care of overlaps
plt.tight_layout()

# axes.plot(x,y)
```

#Figure size and DPI

```python
fig = plt.figure(figsize=(8,4), dpi=100)  #dpi is pixel per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(3,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#save a figure to a file
fig.savefig('mypic.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='quad')
ax.plot(x,x**3, label='cube')

ax.legend(loc=0)    #loc is the location of legend
```

Plot appearances

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='violet', linewidth=3, alpha=0.8, linestyle='dashdot', marker='o', 
        markersize=20, markerfacecolor='purple') #alpha sets the transperancy, linewidth sets the line width

#you can use lw for linewidth and linestyle as ls
```
